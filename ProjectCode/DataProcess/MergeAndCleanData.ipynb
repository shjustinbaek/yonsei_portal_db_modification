```python
#여러개로 분할되어 있는 csv파일 하나의 DataFrame으로 합치기
def merge_tables(lst):
    rtn = pd.read_csv(lst[0],encoding='utf-8')
    print(rtn.shape)
    for t in lst[1:]:
        temp = pd.read_csv(t,encoding='utf-8')
        print(temp.shape)
        rtn = pd.concat([rtn,temp])
    print("final shape: ",rtn.shape)
    return rtn
```

```python
def make_CLASSPROPERTY(row):
    try:
        if row['Property'] == '폐강':
            if len(CLASSPROPERTY.index) == 0:
                CLASSPROPERTY.loc[0] = np.append(row[:-1].values,'폐강')
            else:
                CLASSPROPERTY.loc[len(CLASSPROPERTY)] = np.append(row[:-1].values,'폐강')
        else:
            for i in row['Property']:
                if len(CLASSPROPERTY.index) == 0:
                    CLASSPROPERTY.loc[0] = np.append(row[:-1].values,i)
                else:
                    CLASSPROPERTY.loc[len(CLASSPROPERTY)] = np.append(row[:-1].values,i)
    except:
        print(row)
```

```python
find1 = re.compile(r"^[^(]*")
def make_SUBJECT_BUILDING(row):
    try:
        for i in row['BuildingName'].split('/'):
            i = i.strip()
            if len(SUBJECT_BUILDING.index) == 0:
                if '(' in i:
                    temp2 = i.split('(')
                    for j in range(len(temp2)):
                        SUBJECT_BUILDING.loc[j] = np.append(row[:-1].values,temp2[j].rstrip(')'))
                else:
                    SUBJECT_BUILDING.loc[0] = np.append(row[:-1].values,i)
            else:
                if '(' in i:
                    temp2 = i.split('(')
                    for j in temp2:
                        SUBJECT_BUILDING.loc[len(SUBJECT_BUILDING)] = np.append(row[:-1].values,j.rstrip(')'))
                else:
                    SUBJECT_BUILDING.loc[len(SUBJECT_BUILDING)] = np.append(row[:-1].values,i)
            break
    except:
        print(row)
```
